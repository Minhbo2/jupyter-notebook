```python
# The normal imports
import numpy as np
from numpy.random import randn
import pandas as pd

# Import the stats library from numpy
from scipy import stats

# These are the plotting modules adn libraries we'll use:
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# Command so that plots appear in the iPython Notebook
%matplotlib inline
```

```python
dataset = randn(100)
```

```python
# for combining plots, by default a kde over a histogram is shown
sns.displot(data = dataset, bins = 25, kde = True)
```

```python
# # hist, rug, and kde are all input arguments to turn those plots on or off

sns.displot(data = dataset, rug = True, kind='kde')
```

```python
# WE can also use pandas data objects for this

from pandas import Series

# Create Series form dataset
ser1 = Series(dataset,name='My_DATA')
```

```python
# Plot Series
sns.displot(ser1,bins=25, kde = True)
```
