```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
df_left = DataFrame({
    'key': ['X','Y','Z','X','Y'],
    'data': range(5)
})
df_left
```

```python
df_right = DataFrame({'group_data': [10,20]}, index=['X','Y'])

df_right
```

```python
# notice the key col on left df and index values on right df are share the same overlapping values
# since both df set are "labeled" differently for columns, we specified which col on left table to join on and allow the index use from the right table to join on,
pd.merge(df_left,df_right,left_on='key',right_index=True)
```

# using hierarchy index with merging

```python
df_left_hr = DataFrame({
    'key1':['SF','SF','SF','LA','LA',],
    'key2':[10,20,30,20,30],
    'data_set':np.arange(5.)
})

df_left_hr
```

```python
# hierachy indexes
df_right_hr =  DataFrame(np.arange(10).reshape(5,2),index=[['LA','LA','SF','SF','SF'],[20,10,10,10,20]],columns=['col_1','col_2'])
df_right_hr
```

```python
# merging df with index hierachy
# datasets merged on key1 and key2 for the left table and using indexes of the right table
# notice data of SF/LA indexes and key2 with value of 30 are not included. This is very similar to INNER JOIN of SQL: return/join on values matched in both tables 
pd.merge(df_left_hr,df_right_hr, left_on=['key1','key2'], right_index=True)
```

```python
# python ver of SQL join
df_left.join(df_right)
```
