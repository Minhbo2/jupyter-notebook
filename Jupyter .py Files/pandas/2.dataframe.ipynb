```python
# lib to handle array
import numpy as np

# lib for data label and data frame
import pandas as pd
from pandas import Series, DataFrame
```

```python
# web browsing handling
import webbrowser
website='https://en.wikipedia.org/wiki/NFL_win-loss_records'
webbrowser.open(website)
```

```python
# once you copy the first 5 rows of the table on the site above, pd built in function will grab it and output the same data table 
nfl_frame = pd.read_clipboard()
nfl_frame
```

```python
# grabbing the columns header
nfl_frame.columns
```

```python
# grabbing data from the 'Team' column
nfl_frame.Team
```

```python
# however, you cant do this with column with more than 1 word in the header
nfl_frame.First NFL season
```

```python
# to do the above
nfl_frame['First NFL season']
```

```python
# creating a new data table from the above table
DataFrame(nfl_frame,columns=['Team','First NFL season','Division'])
```

```python
# if creating a new data table with non-existing column or data, pd will give it null values, instead of throwing error
DataFrame(nfl_frame,columns=['Team','First NFL season','Stadium'])
```

```python
# getting rows, if no row specify, head() will returns the first 5 rows, if want n rows, then head(n)
nfl_frame.head()
```

```python
# tail() will returns the last 5 rows, if not specify, if specify nth row, tail(1)
nfl_frame.tail(1) # 1 = last row
```

```python
# grabbing row on a curtain index
nfl_frame.loc[3]
```

```python
# assigning value to the entire column
nfl_frame['Stadium']="Levi's Stadium"
nfl_frame
```

```python
# we can leverage numpy for this as well
nfl_frame['Stadium'] = np.arange(5)
nfl_frame
```

```python
# creating a series for a single column
stadiums = Series(["Levi's Stadium","AT&T"], index=[4,0])
stadiums
```

```python
# pandas built in feature to handle data assignment matches index in the stadiums Series
nfl_frame['Stadium']=stadiums
nfl_frame
```

```python
# we can delete entire column

del nfl_frame['Stadium']
nfl_frame
```
