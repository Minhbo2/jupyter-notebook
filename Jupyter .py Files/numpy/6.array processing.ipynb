```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib.inline
```

```python
points = np.arange(-5,5,0.01)
```

```python
dx,dy = np.meshgrid(points,points)
dx
```

```python
dy
```

```python
z = (np.sin(dx) + np.sin(dy))
z
```

```python
plt.imshow(z)
```

```python
plt.imshow(z)
plt.colorbar()
plt.title('Sin(x) + sin(y)')
```

```python
# numpy where

a = np.array([1,2,3,4])
b = np.array([100,200,300,400])
```

```python
condition = np.array([True,True,False,False])

# list comprehension
answer = [(a_val if cond else b_val) for a_val, b_val, cond in zip(a,b,condition)]
```

```python
# choose values in a arr if condition is True in condition arr, otherwise choose b value in b arr
answer
```

```python
# shorter way to do list comprehension above
answer2 = np.where(condition, a,b)
answer2
```
